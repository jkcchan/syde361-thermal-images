```python
from collections import defaultdict
import numpy
from scipy.stats import norm
import matplotlib
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

variance_hashmap_class3 = defaultdict(list)
variance_hashmap_class2 = defaultdict(list)

with open('regression_3.csv' ,'r') as csv_file:
    lines = csv_file.read().split('\n')
    # start at 1 because first is header
    for i in range(1, len(lines)):
        csv_fields = lines[i].split(',')
        if len(csv_fields) > 1:
            # simple mean of the simple variances
            class_key = csv_fields[-2]
            variance_hashmap_class3[class_key].append(float(csv_fields[-1]))
            variance_hashmap_class2[csv_fields[2]].append(float(csv_fields[-1]))
            
print 'three positions:'
print sum(variance_hashmap_class3['1'])/ len(variance_hashmap_class3['1'])
print sum(variance_hashmap_class3['2'])/ len(variance_hashmap_class3['2'])
print sum(variance_hashmap_class3['3'])/ len(variance_hashmap_class3['3'])

print '\nbinary means:'
print sum(variance_hashmap_class2['0'])/ len(variance_hashmap_class2['0'])
print sum(variance_hashmap_class2['1'])/ len(variance_hashmap_class2['1'])


print '\nbinary std:'
print numpy.std(numpy.array(variance_hashmap_class2['0']), axis=0)
print numpy.std(numpy.array(variance_hashmap_class2['1']), axis=0)



plt.hist(variance_hashmap_class2['1'])
plt.show()
print len(variance_hashmap_class2['1'])
```

```python
plt.hist(variance_hashmap_class2['0'])
plt.show()
print len(variance_hashmap_class2['0'])
```

```python
print(len(variance_hashmap_class2['1']))
# best fit of data
(mu, sigma) = norm.fit(variance_hashmap_class2['1'])

# the histogram of the data
n, bins, patches = plt.hist(variance_hashmap_class2['1'], 50, normed=1, facecolor='blue', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)

#plot
plt.xlabel('Pixel Intensity Variance')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ Pixel\ Intensity\ Variance\ of\ Lying\ Down\ Bounding-Box\:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)

plt.show()
```

```python
print(len(variance_hashmap_class2['0']))
# best fit of data
(mu, sigma) = norm.fit(variance_hashmap_class2['0'])

# the histogram of the data
n, bins, patches = plt.hist(variance_hashmap_class2['0'], 60, normed=1, facecolor='blue', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)

#plot
plt.xlabel('Pixel Intensity Variance')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ Pixel\ Intensity\ Variance\ of\ Non-Lying\ Down\ Bounding-Box\:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)

plt.show()
```
